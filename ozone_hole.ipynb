```python
import os, numpy as np
from facets import facets
from matplotlib import colors
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.colors import Normalize
os.environ['PROJ_LIB'] = '/usr/share/proj'
from mpl_toolkits.basemap import Basemap, addcyclic, cm
```

```python
loc = dict()
loc['Arrival Heights'] = (-78, 166)
loc['Davis'] = (-69, 78)
loc['South Pole'] = (-90, 335)
loc['Neumayer'] = (-71, 352)
loc['Syowa'] = (-69, 40)
loc['Maitri'] = (-70.5, 11.4)
loc['Marambio'] = (-64, 303)
loc['McMurdo'] = (-78, 167)
loc['Concordia'] = (-75, 123)
```

```python
fig, ax = facets(1, 1, width = 8, aspect = 0.5, 
                 top_pad = 0.1, bottom_pad = 0.1, 
                 left_pad = -1.75, right_pad = -1.75)
ax = ax[0]
m = Basemap(projection = 'spstere', lon_0 = 180, boundinglat = -60, round = True)
c = 'black'
size = 20
m.etopo()
# m.shadedrelief()
m.drawcoastlines(linewidth=0.35)
m.drawcountries(linewidth=0.35)
m.drawmeridians(np.arange(0, 360, 30), labels = [0,0,1,0], fontsize = size)
m.drawparallels(np.arange(-90,90, 10), labels = [1,1,0,0], fontsize = size)

# Station Co-ordinates
stations = ['Maitri']
for station in stations: 
    xx, yy = m(loc[station][1], loc[station][0])
    m.scatter(xx, yy, color='r')
    plt.text(xx, yy + 250000, station, color = c, fontsize = size)
plt.savefig('./figs/loc.png', bbox_to_inches='tight')
plt.savefig('./figs/loc.pdf', bbox_to_inches='tight')
plt.show()
```
